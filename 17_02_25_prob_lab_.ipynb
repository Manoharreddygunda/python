```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom
np.random.seed(42)
N = 500
vehicle_counts = np.random.choice([0, 1, 2, 3, 4, 5], size=N, p=[0.1, 0.2, 0.3, 0.2, 0.15, 0.05])
unique, counts = np.unique(vehicle_counts, return_counts=True)
pmf = counts / N
cdf = np.cumsum(pmf)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=unique, y=pmf, color='b', alpha=0.7)
plt.xlabel('Number of Vehicles')
plt.ylabel('PMF')
plt.title('Probability Mass Function')

plt.subplot(1, 2, 2)
plt.plot(unique, cdf, marker='o', linestyle='-', color='r')
plt.xlabel('Number of Vehicles')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function')
plt.show()
```
