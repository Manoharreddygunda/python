```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Question 1: Generate normal distribution data and visualize
mean_q1 = 50
std_dev_q1 = 5
size_q1 = 1000

data_q1 = np.random.normal(mean_q1, std_dev_q1, size_q1)

# Plot histogram
plt.figure(figsize=(10,5))
plt.hist(data_q1, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Histogram')

# Plot Probability Density Function (PDF)
x = np.linspace(mean_q1 - 4*std_dev_q1, mean_q1 + 4*std_dev_q1, 1000)
pdf_q1 = stats.norm.pdf(x, mean_q1, std_dev_q1)
plt.plot(x, pdf_q1, 'r', label='PDF')
plt.title("Normal Distribution: Histogram & PDF")
plt.xlabel("Data Points")
plt.ylabel("Density")
plt.legend()
plt.show()

# Question 2: Probability calculations
mean_q2 = 14
std_dev_q2 = 2.5
size_q2 = 1000

def calculate_probability(lower, upper):
    prob = stats.norm.cdf(upper, mean_q2, std_dev_q2) - stats.norm.cdf(lower, mean_q2, std_dev_q2)
    return prob * size_q2

# I. Students scored between 14 and 2.5
students_between = calculate_probability(2.5, 14)
print(f"Number of students who scored between 14 and 2.5: {students_between:.2f}")

# II. Students scored below 12
students_below_12 = calculate_probability(-np.inf, 12)
print(f"Number of students who scored below 12: {students_below_12:.2f}")

# III. Students scored above 19
students_above_19 = calculate_probability(19, np.inf)
print(f"Number of students who scored above 19: {students_above_19:.2f}")

# Visualization of Normal Distribution with probabilities
x_q2 = np.linspace(mean_q2 - 4*std_dev_q2, mean_q2 + 4*std_dev_q2, 1000)
pdf_q2 = stats.norm.pdf(x_q2, mean_q2, std_dev_q2)
plt.figure(figsize=(10,5))
plt.plot(x_q2, pdf_q2, 'b', label='PDF')
plt.fill_between(x_q2, pdf_q2, where=(x_q2 >= 2.5) & (x_q2 <= 14), color='g', alpha=0.5, label='Between 2.5 and 14')
plt.fill_between(x_q2, pdf_q2, where=(x_q2 <= 12), color='r', alpha=0.5, label='Below 12')
plt.fill_between(x_q2, pdf_q2, where=(x_q2 >= 19), color='y', alpha=0.5, label='Above 19')
plt.title("Normal Distribution with Marked Probability Areas")
plt.xlabel("Scores")
plt.ylabel("Density")
plt.legend()
plt.show()
```
