```python
import matplotlib.pyplot as plt

# Data
study_hours = [2, 4, 6, 8, 1, 5, 7, 3, 9, 10]
exam_scores = [50, 60, 70, 85, 40, 65, 78, 55, 90, 95]
sleep_hours = [8, 7, 6, 5, 9, 6.5, 5.5, 7.5, 4, 3]


plt.figure(figsize=(5,4))
plt.scatter(study_hours, exam_scores, color='blue')
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study Hours vs. Exam Scores")
plt.grid(True)
plt.show()


plt.figure(figsize=(5,4))
plt.scatter(sleep_hours, exam_scores, color='red')
plt.xlabel("Sleep Hours")
plt.ylabel("Exam Scores")
plt.title("Sleep Hours vs. Exam Scores")
plt.grid(True)
plt.show()


plt.figure(figsize=(5,4))
plt.scatter(study_hours, sleep_hours, color='green')
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Study Hours vs. Sleep Hours")
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Calculate Pearson correlation
corr_study_exam = np.corrcoef(study_hours, exam_scores)[0, 1]
corr_sleep_exam = np.corrcoef(sleep_hours, exam_scores)[0, 1]
corr_study_sleep = np.corrcoef(study_hours, sleep_hours)[0, 1]

# Print correlation values
print(f"Pearson Correlation (Study Hours vs. Exam Scores): {corr_study_exam:.2f}")
print(f"Pearson Correlation (Sleep Hours vs. Exam Scores): {corr_sleep_exam:.2f}")
print(f"Pearson Correlation (Study Hours vs. Sleep Hours): {corr_study_sleep:.2f}")
```
